```python
#Anova test
from scipy.stats import stats
F, p = stats.f_oneway(df[df["species"] == "setosa"]["sepal_length"],df[df["species"] == "versicolor"]["sepal_length"],df[df["species"] == "virginica"]["sepal_length"])
print("p-value for significance is: ", p)
```

```python
#p is less than 5% so reject null Hypothesis
```

```python
#Anova test2

F2, p2 = stats.f_oneway(df[df["species"] == "setosa"]["sepal_width"],df[df["species"] == "versicolor"]["sepal_width"],df[df["species"] == "virginica"]["sepal_width"])
print("p-value for significance is: ", p2)
```

```python
#p is less than 5% so reject null Hypothesis
```
